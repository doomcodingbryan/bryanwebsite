### Key Terms and Definitions

**Key Terms:**
- **ZORI (Zillow Observed Rent Index)**: A smoothed measure of the typical observed market rate rent across a given region, representing the median rent for all homes in the 40th to 60th percentile range.
- **Student-to-Resident Ratio**: The proportion of enrolled university students relative to the total city population, expressed as a percentage.
- **CAGR (Compound Annual Growth Rate)**: The mean annual growth rate of rent over a specified time period, calculated to show the smoothed rate of change.
- **Rent Volatility**: The standard deviation of month-to-month percentage changes in rent prices, measuring the variability and instability of rental markets.
- **Housing Unit Growth Rate**: The percentage change in total housing units over a specified period, indicating the rate at which new housing supply is added to the market.
- **College Town**: A city where a significant portion of the population consists of university students, typically characterized by student-oriented housing and businesses.
- **Studentification**: The process by which an increasing number of students concentrate in particular neighborhoods, transforming the social, cultural, and physical environment of that area.
- **Supply Constraint**: Limitations on housing construction due to zoning regulations, land availability, or development restrictions that prevent housing supply from meeting demand.

### Hypothesis 1: Student-to-Resident Ratio and Rent Prices <a name="hyp1-analysis"></a>

**Hypothesis Statement:** College towns with a higher student-to-resident ratio have higher average rents due to concentrated demand pressure on a thin local market.

**Methodology:** We run a linear regression where the dependent variable is average rent price and the independent variable is the student-to-resident population ratio. If β₁ > 0 and statistically significant, this would support the hypothesis that a higher proportion of students in a city is associated with higher average rent prices.

**Model:** avg_rent = β₀ + β₁(student_ratio) + ε

We load the cleaned and merged dataset. This dataset contains:

* College and city information
* Average rent prices for each city
* Student-to-resident ratios (as decimals)
* State abbreviations for analysis

```python
regression_df = pd.read_csv('student_rent_analysis.csv')
print(regression_df.head())
```

Before running the regression, we examine the descriptive statistics of our key variables: average rent and student-to-resident ratio.

```python
print("Descriptive Statistics:")
print(regression_df[['avg_rent', 'student_ratio']].describe())
```

We fit an OLS (Ordinary Least Squares) regression model with:

* **Dependent variable (Y):** Average rent price
* **Independent variable (X):** Student-to-resident ratio
* **Model:** avg_rent = β₀ + β₁(student_ratio) + ε

```python
X = regression_df[['student_ratio']]
y = regression_df['avg_rent']
X_with_const = sm.add_constant(X)
model = sm.OLS(y, X_with_const).fit()
print("\n" + model.summary().as_text())
```

We extract the important coefficients and statistics from the regression model to clearly present the findings.

```python
beta_0, beta_1 = model.params[['const', 'student_ratio']]
p_value = model.pvalues['student_ratio']
r2, r2_adj = model.rsquared, model.rsquared_adj
ci_low, ci_high = model.conf_int().loc['student_ratio']

print("KEY RESULTS")
print("=" * 70)

print(f"\nRegression Equation:\navg_rent = {beta_0:.2f} + {beta_1:.2f} * student_ratio")

print(f"\nβ₀ (intercept): ${beta_0:.2f}")
print(f"β₁ (coefficient for student_ratio): ${beta_1:.2f}")
print(f"p-value for β₁: {p_value:.6f}")
print(f"R-squared: {r2:.4f}")
print(f"Adjusted R-squared: {r2_adj:.4f}")

print(f"95% CI for β₁: [${ci_low:.2f}, ${ci_high:.2f}]")
```

We create four diagnostic plots to assess the regression model:

```python
# Scatterplot with regression line
fig, ax = plt.subplots(figsize=(7, 5))

ax.scatter(
    regression_df['student_ratio'], regression_df['avg_rent'],
    alpha=0.6, edgecolor='k'
)

ax.plot(
    regression_df['student_ratio'], model.fittedvalues,
    linewidth=2,
    label=rf'$\beta_1 = {beta_1:.2f},\ p = {p_value:.4f}$'
)

ax.set_xlabel('Student-to-Resident Ratio')
ax.set_ylabel('Average Rent Price ($)')
ax.set_title('Rent Price vs Student-to-Resident Ratio')

ax.legend()
ax.grid(alpha=0.3)

plt.show()
```

This scatterplot illustrates the relationship between student-to-resident ratio and average rent prices across 81 college towns. Each point represents a town, with the x-axis showing the proportion of students and the y-axis showing average rent in dollars. The regression line represents the predicted relationship, with the slope coefficient (β₁ = $386.43) and its p-value (0.3568) displayed in the legend. The nearly flat line and high p-value indicate a weak and statistically insignificant relationship, while the wide spread of points shows that towns with similar student ratios can have very different rents.

```python
# Residual Plot
fig, ax = plt.subplots(figsize=(7, 5))

residuals = model.resid

ax.scatter(model.fittedvalues, residuals, alpha=0.6, edgecolor='k')
ax.axhline(0, color='red', linestyle='--', linewidth=2)

ax.set_xlabel('Fitted Values')
ax.set_ylabel('Residuals')
ax.set_title('Residual Plot')

ax.grid(alpha=0.3)

plt.show()
```

This residual plot shows the residuals from the regression model plotted against the predicted rent values, allowing assessment of the model’s assumptions. The x-axis represents the fitted values, while the y-axis shows the residuals (actual minus predicted rent), and the red dashed line indicates zero residuals. The residuals are scattered randomly around zero with no obvious pattern, suggesting that a linear model is appropriate and that the errors are roughly consistent across rent levels (homoscedasticity). Some outliers with large positive residuals, likely representing high-rent cities such as Malibu or New York, highlight cases where actual rent far exceeded predictions. Overall, the random scatter supports the linearity assumption, though these extreme values indicate limitations in the model’s predictive power.

### Hypothesis 2: Rent Volatility and Long-Term Growth <a name="hyp2-analysis"></a>

**Hypothesis Statement:** Among college towns, greater month-to-month rent volatility is associated with higher long-term rent growth.

**Methodology:** For each town we compute volatility as the standard deviation of the month-to-month percentage changes in the Zillow ZORI data and compute long-term growth as the compounded annual growth rate (CAGR). Then we run a linear regression where the dependent variable is CAGR and the independent variable is volatility. If β_volatility > 0 and statistically significant, this would support the hypothesis that higher rent volatility is associated with higher long-term growth.

**Model:** CAGR = β₀ + β₁(volatility) + ε

We load the pre-cleaned datasets from the Data Cleaning section that already contain City_State identifiers and month-to-month rent percentage changes.

```python
# Load pre-cleaned datasets from Data Cleaning section
zillow_df = pd.read_csv('clean_zillow_with_pctchange.csv')
colleges_df = pd.read_csv('clean_colleges_with_abbrev.csv')

# Convert Date back to datetime (CSVs store as strings)
zillow_df['Date'] = pd.to_datetime(zillow_df['Date'])

print(
    f"Loaded {len(zillow_df)} Zillow observations across "
    f"{zillow_df['City_State'].nunique()} cities"
)
print(f"Loaded {len(colleges_df)} college towns")
```

We compute volatility as the standard deviation of monthly percentage changes for each city, and CAGR to measure long-term growth.

```python
# Remove missing values from percentage change calculations
zillow_pct = zillow_df.dropna(subset=['Rent_PctChange'])

# Calculate volatility (std dev of monthly % changes)
volatility_df = zillow_pct.groupby('City_State').agg({
    'Rent_PctChange': ['std', 'count']
}).reset_index()

volatility_df.columns = ['City_State', 'Volatility', 'N_Months']

print("\nSample cities:")
print(volatility_df.head(10))
```

We calculate the Compound Annual Growth Rate (CAGR) for each city to measure long-term rent growth. CAGR smooths out short-term fluctuations and shows the average annual growth rate over the time period.

```python
# Calculate CAGR (Compound Annual Growth Rate)
# CAGR = (Ending Value / Beginning Value)^(1/years) - 1
cagr_df = zillow_df.groupby('City_State').agg({
    'RentIndex': ['first', 'last'],
    'Date': ['min', 'max']
}).reset_index()

cagr_df.columns = ['City_State', 'Start_Rent', 'End_Rent', 
                   'Start_Date', 'End_Date']

# Calculate years elapsed
cagr_df['Years'] = (cagr_df['End_Date'] - cagr_df[
            'Start_Date']).dt.days / 365.25

# Calculate CAGR
cagr_df['CAGR'] = ((cagr_df['End_Rent'] / cagr_df[
    'Start_Rent']) ** (1 / cagr_df['Years']) - 1) * 100

print("\nSample cities:")
print(cagr_df[['City_State', 'Start_Rent',
                'End_Rent', 'Years', 'CAGR']].head(10))
```

We merge volatility and CAGR data and filter for cities with at least 24 months of observations to ensure reliable measurements.

```python
# Merge volatility and CAGR data
analysis_df = volatility_df.merge(cagr_df[['City_State', 'CAGR', 'Years']], on='City_State')

# Filter for cities with sufficient data (at least 24 months of data)
analysis_df = analysis_df[analysis_df['N_Months'] >= 24]

print(f"Analysis dataset: {len(analysis_df)} cities with sufficient data")
print("\nCombined data sample:")
print(analysis_df.head(10))
```

We identify which cities in our analysis are college towns and merge in the student population data.

```python
# Get the list of college town City_State identifiers
college_towns_list = colleges_df['City_State'].tolist()

# Add a flag to indicate if a city is a college town
analysis_df['Is_College_Town'] = analysis_df['City_State'].isin(college_towns_list)

# Merge with college data to get Student_City_Ratio
analysis_df = analysis_df.merge(
    colleges_df[['City_State', 'Student_City_Ratio', 'Institution', 'Undergrad_Population']], 
    on='City_State', 
    how='left'
)

print(f"\nTotal cities in analysis: {len(analysis_df)}")
print(f"College towns in analysis: {analysis_df['Is_College_Town'].sum()}")
print(f"Non-college towns in analysis: {(~analysis_df['Is_College_Town']).sum()}")

print("\nCollege towns in our dataset:")
college_analysis = analysis_df[analysis_df['Is_College_Town']].copy()
print(college_analysis[['City_State', 'Student_City_Ratio', 
                'Volatility', 'CAGR']].sort_values('Student_City_Ratio', ascending=False))
```

Before running the regression, we examine the descriptive statistics and correlation between our key variables: volatility and CAGR.

```python
# Filter for college towns only
college_towns_analysis = analysis_df[analysis_df['Is_College_Town']].copy()

print("Descriptive Statistics (College Towns Only):")
print(college_towns_analysis[['Volatility', 'CAGR']].describe())

print(f"\nNumber of college towns in analysis: {len(college_towns_analysis)}")
```

We fit an OLS (Ordinary Least Squares) regression model with:
- **Dependent variable (Y):** CAGR (long-term rent growth)
- **Independent variable (X):** Volatility (standard deviation of monthly rent changes)
- **Model:** CAGR = β₀ + β₁(volatility) + ε

```python
X = college_towns_analysis[['Volatility']]
y = college_towns_analysis['CAGR']
X_with_const = sm.add_constant(X)
model = sm.OLS(y, X_with_const).fit()
print("\n" + model.summary().as_text())
```

```python
beta_0, beta_1 = model.params[['const', 'Volatility']]
p_value = model.pvalues['Volatility']
r2, r2_adj = model.rsquared, model.rsquared_adj
ci_low, ci_high = model.conf_int().loc['Volatility']

print("KEY RESULTS")
print("=" * 70)

print(f"\nRegression Equation:\nCAGR = {beta_0:.4f} + {beta_1:.4f} * volatility")

print(f"\nβ₀ (intercept): {beta_0:.4f}")
print(f"β₁ (coefficient for volatility): {beta_1:.4f}")
print(f"p-value for β₁: {p_value:.6f}")
print(f"R-squared: {r2:.4f}")
print(f"Adjusted R-squared: {r2_adj:.4f}")

print(f"95% CI for β₁: [{ci_low:.4f}, {ci_high:.4f}]")
```

We create four diagnostic plots to assess the regression model:

```python
# Scatterplot with regression line
fig, ax = plt.subplots(figsize=(7, 5))

ax.scatter(
    college_towns_analysis['Volatility'], college_towns_analysis['CAGR'],
    alpha=0.6, edgecolor='k'
)

ax.plot(
    college_towns_analysis['Volatility'], model.fittedvalues,
    linewidth=2,
    label=rf'$\beta_1 = {beta_1:.4f},\ p = {p_value:.4f}$'
)

ax.set_xlabel('Rent Volatility (Std Dev of Monthly % Changes)')
ax.set_ylabel('Long-term Rent Growth (CAGR %)')
ax.set_title('CAGR vs Rent Volatility')

ax.legend()
ax.grid(alpha=0.3)

plt.show()
```

The scatterplot reveals no clear relationship between rent volatility and long-term growth. The nearly flat regression line (β₁ = -0.0971, p = 0.8308) and widely scattered points show that college towns with high month-to-month volatility do not consistently experience higher CAGR. Some cities have stable rents but strong growth, while others show high volatility without corresponding long-term gains, indicating these two metrics are driven by different market forces.

```python
print("NOTABLE CITIES")
print("=" * 70)

def print_cities(title, df):
    print(f"\n{title}:")
    for _, row in df.iterrows():
        print(f"  {row['City_State']}: "
              f"Volatility={row['Volatility']:.3f}, "
              f"CAGR={row['CAGR']:.2f}%")

print_cities(
    "Top 5 cities by volatility",
    college_towns_analysis.nlargest(5, 'Volatility')[['City_State', 'Volatility', 'CAGR']]
)

print_cities(
    "Top 5 cities by CAGR (long-term growth)",
    college_towns_analysis.nlargest(5, 'CAGR')[['City_State', 'Volatility', 'CAGR']]
)

print_cities(
    "Bottom 5 cities by CAGR",
    college_towns_analysis.nsmallest(5, 'CAGR')[['City_State', 'Volatility', 'CAGR']]
)
```

The top cities by volatility show varied CAGR values, while the highest growth cities have moderate volatility. Cities with negative growth also display diverse volatility levels. This pattern confirms that rent volatility and long-term growth are not strongly correlated in college town markets.

```python
# Histogram
fig, ax = plt.subplots(figsize=(7, 5))

residuals = model.resid

ax.hist(residuals, bins=20, edgecolor='black', alpha=0.7)
ax.axvline(0, color='red', linestyle='--', linewidth=2)

ax.set_xlabel('Residuals')
ax.set_ylabel('Frequency')
ax.set_title('Distribution of Residuals')
ax.grid(alpha=0.3, axis='y')

plt.show()
```

The histogram shows residuals are roughly bell-shaped and centered near zero, consistent with the normality assumption. However, the relatively wide spread (reflecting R² = 0.001) confirms that volatility alone poorly predicts long-term rent growth across college towns.

```python
# Q-Q Plot
fig, ax = plt.subplots(figsize=(7, 5))

sm.qqplot(model.resid, line='45', ax=ax)

ax.set_title('Q–Q Plot (Normality Check)')
ax.grid(alpha=0.3)

plt.show()
```

The Q-Q plot shows residuals generally follow the diagonal line, indicating approximate normality. Slight deviations in the tails suggest a few outlier college towns with unusually high or low growth rates relative to their volatility. Overall, the residuals are reasonably normal, supporting the validity of the regression's statistical inference.

```python
# Residual Plot
fig, ax = plt.subplots(figsize=(7, 5))

residuals = model.resid

ax.scatter(model.fittedvalues, residuals, alpha=0.6, edgecolor='k')
ax.axhline(0, color='red', linestyle='--', linewidth=2)

ax.set_xlabel('Fitted Values')
ax.set_ylabel('Residuals')
ax.set_title('Residual Plot')

ax.grid(alpha=0.3)

plt.show()
```

The residual plot displays random scatter around zero with no obvious pattern, indicating the linear model is appropriate. The consistent spread across fitted values suggests homoscedasticity (constant error variance). While this validates the model assumptions, the low R² indicates that factors beyond volatility drive most of the variation in long-term rent growth.

<h4> Interpretation </h4>

The regression results indicate that the relationship between rent volatility and long-term rent growth in college towns is extremely weak. The estimated coefficient for volatility is slightly negative (β = –0.0971), meaning that towns with higher month-to-month fluctuations in rent do not experience faster long-term growth—in fact, they exhibit marginally lower growth on average. However, this effect is not statistically significant (p = 0.8308), well above the α = 0.05 threshold, meaning we cannot conclude that this effect is real rather than just random chance.

Additionally, the model's R-squared value is extremely low (R² = 0.001), indicating that volatility explains only about 0.1% of the variation in long-term rent growth. This suggests that factors other than short-term rent fluctuations are driving most of the differences in long-term growth across college towns.

**Model Oddities:** Several unexpected patterns emerged in our analysis. First, some college towns with the highest volatility actually experienced below-average long-term growth, contradicting our hypothesis that volatile markets signal high demand. Second, cities with negative CAGR show a wide range of volatility levels, suggesting that market decline can occur regardless of whether prices fluctuate wildly or remain relatively stable month-to-month. Third, our linear model treats volatility symmetrically. It assumes that an increase in volatility has the same effect regardless of the starting level. However, the data suggest that moderate volatility may be normal in college towns due to seasonal leasing cycles, while extreme volatility could indicate different underlying problems entirely. Finally, the near-zero R² is itself unusual, as it suggests volatility and growth are essentially unrelated in college towns, which contradicts urban economics literature suggesting that supply constraints manifest in both metrics simultaneously.

Based on these results, we fail to reject the null hypothesis. Although the estimated coefficient is positive, with a p-value of 0.8264 and an R² of just 0.001, there is no reliable evidence that higher rent volatility is associated with higher long-term rent growth in college towns. 

### Hypothesis 3: Housing Unit Growth and Rent Increases <a name="hyp3-analysis"></a>

**Hypothesis Statement:** College towns with lower housing unit growth rates experience faster rent increases than those of higher housing unit growth rates.

**Methodology:** Using the dataset of top college towns, we calculated the Housing Unit Growth Rate for each town as the percentage change in housing units from 2020 to 2024. We then ran a linear regression with Rent Growth as the dependent variable and Housing Growth as the independent variable. A negative coefficient ($\beta < 0$) would indicate that as housing supply growth increases, rent growth decreases, supporting the hypothesis.

Note: Due to data limitations regarding median income and renter share in the available files, we compared college towns against the full set of non-college towns in our dataset rather than a propensity-matched set. This provides a broader market baseline for comparison.

**Model:** rent_growth = β₀ + β₁(housing_growth) + ε

We load the merged college town dataset that contains housing unit data and rent growth information.

```python
merged_ct_data = pd.read_csv("merged_collegetown_data.csv")

print(merged_ct_data.head())
```

Because we cannot compare raw numbers as big cities naturally build more units than small towns, we calculate the percentage growth and normalize the data, allowing us to compare the rate of construction in a small college town versus a large one. By using the housing unit data in 2024 and 2020, we can calculate the total housing growth of each college town.

```python
#calculate housing growth
merged_ct_data['Housing_Growth'] = (
    merged_ct_data['2024'] - merged_ct_data['2020']
    ) / merged_ct_data['2020']
```

We run an Ordinary Least Squares (OLS) regression where Rent_Growth (Y) is the dependent variable and Housing_Growth (X) is the independent variable. This tests our hypothesis and helps determine if Housing_Growth is statistically significant. The regression calculates a coefficient (β) that estimates how much rent growth changes for every 1% increase in housing supply. Here, we specifically look for whether β < 0.

```python
Y = merged_ct_data['Rent_Growth']
X = merged_ct_data['Housing_Growth']
X = sm.add_constant(X)

model = sm.OLS(Y, X).fit()

print("="*60)
print("HYPOTHESIS 3: REGRESSION RESULTS (COLLEGE TOWNS)")
print("="*60)
print(model.summary())
```

We visualize the data to show the relationship between housing growth and rent growth. This scatterplot and regression line helps us see if the linear model makes sense or if the relationship is driven by outliers.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(merged_ct_data['Housing_Growth'], 
            merged_ct_data['Rent_Growth'], 
            alpha=0.7, color='blue', 
            label='College Towns')
plt.plot(merged_ct_data['Housing_Growth'], 
         model.predict(X), 
         color='red', linewidth=2, 
         label=f'Trend Line (β={
             model.params["Housing_Growth"]:.4f})')
plt.xlabel('Housing Growth')
plt.ylabel('Rent Growth')
plt.title('Rent Growth vs. Housing Growth in College Towns')
plt.legend()
plt.show()
```

<h4> Interpretation </h4>

The regression analysis yields a negative coefficient (β = -0.0687), which directionally aligns with the hypothesis – college towns with higher housing supply growth tend to see slightly lower rent increases. Specifically, for every 1% increase in housing supply, annual rent growth is lower by approximately 0.07%.

However, we conclude that this relationship is not statistically significant (p = 0.342 > 0.05). The R-squared value = 0.012 is very low, showing that housing supply growth explains only about 1% of the variation in rent prices across college towns. This suggests that for this specific set of college towns during this time period (2020-2024), rent increases were likely driven by factors other than local housing supply constraints, such as demand from student populations or inflation.

In terms of oddities, the model assumes that the relationship between housing and rent is unbounded and constant. Theoretically, the model suggests and allows that with enough housing growth, the rent price or growth can turn negative. However this is realistically and economically impossible, as rent usually have floor prices, determined by operating costs and taxes, therefore it can never go negative. So in real life, if there was a sudden boom in housing growth, supply-price relationship would likely flatten out rather than continue downward linearly forever. The model also allows for a sudden, unbounded, and significant decrease in housing growth, which is usually unlikely in real life. This phenonemon could technically be possible through natural or man-made disasters – however it should be noted that this instance would be extremely rare.

<h2 id="limitations"> Limitations </h2>

### Data Limitations

**1. Geographic and Temporal Scope**
- Our analysis focuses on U.S. cities with available Zillow rental data from 2019-2025. Cities without consistent Zillow coverage are excluded, which may introduce selection bias toward larger, more urban markets.
- The time period (2019-2025) includes the COVID-19 pandemic, which caused unprecedented disruptions to rental markets, student enrollment patterns, and migration trends. These extraordinary circumstances may not reflect typical market dynamics.

**2. Student Population Data**
- Student enrollment figures represent a single point in time and may not capture year-to-year fluctuations in university populations.
- We focus on undergraduate enrollment at top 100 universities, excluding community colleges, graduate students, and smaller institutions. This may underestimate the true student presence in some cities.
- The student-to-resident ratio assumes all enrolled students live in the host city, which may not be accurate for commuter schools or universities with substantial online enrollment.

**3. Housing Market Complexity**
- The Zillow Observed Rent Index (ZORI) represents median market rents but may not fully capture the student-specific housing market, particularly purpose-built student housing, university-owned dorms, or off-market arrangements.
- Rental markets are influenced by numerous confounding variables not controlled for in our simple regressions: local employment rates, proximity to major metro areas, climate, quality of life factors, and local housing regulations.
- Our analysis treats all college towns as comparable, but university characteristics (public vs. private, research vs. liberal arts, commuter vs. residential) likely create heterogeneous effects on local housing markets.

**4. Causality and Omitted Variable Bias**
- Our regression analyses identify correlations but cannot establish causation. We cannot definitively conclude that student populations *cause* rent changes, as reverse causality or omitted variables may drive the observed relationships.
- Key omitted variables include: local zoning policies, availability of public transportation, regional economic growth, proximity to employment centers, and historical housing development patterns.

**5. Measurement Issues**
- Housing unit growth data is measured at the county level, not the city level, which may introduce noise when colleges are located in large or diverse counties.
- Rent volatility measured as standard deviation of monthly percentage changes may be sensitive to outliers or data quality issues in Zillow's coverage.


### Methodological Limitations

**1. Model Simplicity**
- We primarily use simple linear regression with one or two predictors. More sophisticated models (multiple regression, fixed effects, matching methods) could better isolate the effects of student populations while controlling for confounders.
- The linear functional form may not capture non-linear relationships (e.g., threshold effects where student populations only affect rents above a certain concentration).

**2. Sample Size and Power**
- Our sample size for college towns is limited by the intersection of three datasets (colleges, Zillow coverage, housing unit data). Small sample sizes reduce statistical power and increase the risk of Type II errors (failing to detect real effects).
- Some hypotheses, particularly those involving subgroup analyses (e.g., college towns only), have fewer observations and wider confidence intervals.

**3. External Validity**
- Findings may not generalize beyond the U.S. context, as other countries have different higher education systems, housing markets, and urban development patterns.
- Results are most applicable to traditional college towns with residential university populations and may not apply to cities with primarily commuter schools or online institutions.

### Implications for Interpretation

The limitations of our analysis mean that the results should be interpreted as exploratory and descriptive rather than definitive causal evidence. Any policy recommendations drawn from these findings should be made cautiously and supported with additional local data and context. While the weak or null findings in our hypothesis tests may indicate a true absence of effects, they could also stem from measurement error, omitted variables, or issues in model specification. Our comparative interpretations, such as examining differences between college towns and non-college towns, are more reliable than estimates of absolute effect sizes.

Overall, we can describe general patterns in how student populations relate to rent levels across U.S. cities, identify places that deviate from expected patterns, and compare metrics like volatility or growth rates between city types. However, we cannot make causal claims about student populations influencing rent prices without addressing confounding variables, nor can we accurately predict future rent changes based solely on student ratios or generalize these findings to cities or universities not included in our sample.
